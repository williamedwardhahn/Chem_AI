## Variables — Recording Observations

A **variable** stores a value. Think of it as labeling a flask in the lab.

```python
# Text values are "strings" — wrap them in quotes
molecule_name = "Aspirin"
formula = "C9H8O4"
smiles = "CC(=O)Oc1ccccc1C(=O)O"

# Numbers without decimals are "integers"
num_atoms = 21
cid = 2244  # PubChem Compound ID

# Numbers with decimals are "floats"
molecular_weight = 180.16
logp = 1.24           # partition coefficient (lipophilicity)
melting_point = 135.0  # degrees Celsius

print("Molecule:", molecule_name)
print("Formula:", formula)
print("SMILES:", smiles)
print("Molecular Weight:", molecular_weight, "g/mol")
print("LogP:", logp)
print("Melting Point:", melting_point, "°C")
```

## Lists — Your Compound Library

A **list** holds multiple values in order, like a tray of samples.

```python
# Five common organic molecules
molecules = ["Aspirin", "Caffeine", "Ethanol", "Glucose", "Benzene"]
mol_weights = [180.16, 194.19, 46.07, 180.16, 78.11]

print("Number of compounds:", len(molecules))
print("First compound:", molecules[0])
print("Heaviest:", molecules[mol_weights.index(max(mol_weights))])

# Add a new compound
molecules.append("Acetone")
mol_weights.append(58.08)
print("Updated library:", molecules)
```

## Dictionaries — Compound Data Cards

A **dictionary** maps keys to values, like fields on a safety data sheet.

```python
aspirin = {
    "name": "Aspirin",
    "formula": "C9H8O4",
    "smiles": "CC(=O)Oc1ccccc1C(=O)O",
    "mw": 180.16,
    "logp": 1.24,
    "class": "Acid"
}

print(aspirin["name"], "-", aspirin["formula"])
print("Molecular Weight:", aspirin["mw"], "g/mol")
print("Drug class:", aspirin["class"])
```

```python
# A compound library as a list of dictionaries
library = [
    {"name": "Aspirin",   "mw": 180.16, "logp": 1.24, "class": "Acid"},
    {"name": "Caffeine",  "mw": 194.19, "logp": -0.07, "class": "Alkaloid"},
    {"name": "Ethanol",   "mw": 46.07,  "logp": -0.31, "class": "Alcohol"},
    {"name": "Glucose",   "mw": 180.16, "logp": -3.24, "class": "Sugar"},
    {"name": "Benzene",   "mw": 78.11,  "logp": 2.13, "class": "Aromatic"},
    {"name": "Acetone",   "mw": 58.08,  "logp": -0.24, "class": "Ketone"},
]

for mol in library:
    print(f"{mol['name']:12s}  MW={mol['mw']:7.2f}  LogP={mol['logp']:6.2f}  ({mol['class']})")
```

## Loops — Processing Multiple Compounds

A **loop** repeats code for every item in a list.

```python
molecules = ["Aspirin", "Caffeine", "Ethanol", "Glucose", "Benzene", "Acetone"]
mol_weights = [180.16, 194.19, 46.07, 180.16, 78.11, 58.08]

for name, mw in zip(molecules, mol_weights):
    print(f"{name}: {mw} g/mol")
```

## Conditionals — Classifying Compounds

Use **if/elif/else** to make decisions based on data.

```python
def classify_by_logp(logp):
    """Classify a compound's lipophilicity by its LogP value."""
    if logp > 3:
        return "Very Lipophilic (fat-soluble)"
    elif logp > 0:
        return "Moderately Lipophilic"
    elif logp > -2:
        return "Slightly Hydrophilic"
    else:
        return "Very Hydrophilic (water-soluble)"

for mol in library:
    cat = classify_by_logp(mol["logp"])
    print(f"{mol['name']:12s}  LogP={mol['logp']:6.2f}  -> {cat}")
```

## Functions — Reusable Lab Tools

```python
def daltons_to_kda(mw):
    """Convert molecular weight from Daltons to kiloDaltons."""
    return mw / 1000.0

def drug_likeness(mw, logp, hbd=0, hba=0):
    """Check Lipinski's Rule of Five (simplified)."""
    violations = 0
    if mw > 500: violations += 1
    if logp > 5: violations += 1
    if hbd > 5: violations += 1
    if hba > 10: violations += 1
    return violations

# Test the functions
for mol in library:
    v = drug_likeness(mol["mw"], mol["logp"])
    status = "PASS" if v == 0 else f"FAIL ({v} violations)"
    print(f"{mol['name']:12s}  Lipinski: {status}")
```

## Your First Plot

Let's visualize our compound library.

```python
import matplotlib.pyplot as plt

names = [m["name"] for m in library]
weights = [m["mw"] for m in library]
logps = [m["logp"] for m in library]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Molecular weight bar chart
colors = ["#e74c3c", "#3498db", "#2ecc71", "#f39c12", "#9b59b6", "#1abc9c"]
ax1.bar(names, weights, color=colors)
ax1.set_title("Molecular Weight by Compound")
ax1.set_ylabel("Molecular Weight (g/mol)")
ax1.tick_params(axis="x", rotation=30)
ax1.grid(axis="y", alpha=0.3)

# LogP scatter
ax2.scatter(weights, logps, c=colors, s=100, edgecolors="black", zorder=3)
for i, name in enumerate(names):
    ax2.annotate(name, (weights[i], logps[i]), textcoords="offset points", xytext=(6, 6), fontsize=9)
ax2.set_title("Molecular Weight vs. LogP")
ax2.set_xlabel("Molecular Weight (g/mol)")
ax2.set_ylabel("LogP")
ax2.axhline(0, color="gray", linestyle="--", alpha=0.5)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Compounds above the dashed line are lipophilic (fat-soluble).")
print("Compounds below are hydrophilic (water-soluble).")
```
